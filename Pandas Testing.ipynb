# Group By Column

```python
#Get Mean/Standard Deviation of # of units per district
dfg = df[['district','n_units']].groupby('district').agg([np.mean, np.std, np.size])
dfg
```

```python
#Access Grouped statistic
fin_mean = dfg.ix['FINANCIAL']['n_units']['mean']
fin_std = dfg.ix['FINANCIAL']['n_units']['std']
print "Mean # of Units in Financial District is %s with %s StDev." % (fin_mean, fin_std)
```

# Group By Index

# Merge, Join, and Concatenate

### Merge

If we try to do this on df, which is indexed on property_id instead of district, the n_units_mean column is filled with NaN's.

# Creating new variables/columns

```python
#Create a constant variable for Linear Regression
df['const'] = 1

#Compute statistics by district
dfg = df[['district','n_units']].groupby('district').agg([np.mean, np.std, np.size])
dfg = dfg['n_units'].rename(columns={'mean': 'n_units_mean', 'std': 'n_units_std', 'size': 'n_properties'})

#Impute Missing values
n_units_mean = df['n_units'].mean()
df['n_units'].fillna(n_units_mean, inplace=True)

dfg[:3]
```

```python
df[:5]
```

```python
#Normalize the n_units via standard normal transformation
dfn = dfx.join(dfg)
dfn['n_units_std']

dfn['nrm_n_units'] = (dfn['n_units'] - dfn['n_units_mean'])/dfn['n_units_std']
 
dfn[['district', 'n_units', 'n_units_mean', 'nrm_n_units']].ix['ALPHABET CITY'][:5]
```

# Create variables with apply

### Numpy Functions

```python
#Apply square root to n units
dfn['sqrt_n_units'] = dfn['n_units'].apply(np.sqrt)
dfn[['district', 'n_units', 'n_units_mean', 'nrm_n_units','sqrt_n_units']][:5]
```

### Custom Functions

```python
#Let's take a look at the distribution real quick to determine cut-offs
dfn['n_units'].describe()
```

```python
#Define a function
def map_n_units_catgy(x, min, max):
    if x <= min:
        return 'Low'
    elif min < x < max:
        return 'Medium'
    elif x > max:
        return 'High'

#apply the function to our dataframe
df['catgy_n_units'] = df['n_units'].apply(map_n_units_catgy, args=(23,120))
```

```python
df['catgy_n_units'][:5]
```
