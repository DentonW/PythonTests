```python
%pylab inline
import pandas as pd
from ggplot import *
```

```python
# Gets rid of the giant minus signs that matplotlib likes
mpl.rcParams['axes.unicode_minus'] = False
rcdef = plt.rcParams.copy()
```

### Creating the Example Data Sets

```python
df1 = pd.Series(np.random.randn(100))
df1 = df1.cumsum()
df2 = pd.Series(np.random.randn(100))
df2 = df2.cumsum()

# Combine into a single dataframe (could have done this in one step)
df = pd.DataFrame({'x': pd.Series(range(100)), 'Run 1': df1, 'Run 2': df2})
# Convert to lists for simple matplotlib
run1 = df["Run 1"].tolist()
run2 = df["Run 2"].tolist()
xaxis = range(100)

df.describe()
```

### Plot in Matplotlib

This is your standard matplotlib plot.

```python
pyplot.plot(xaxis, run1, label='Run 1')
pyplot.plot(xaxis, run2, label='Run 2')
pyplot.legend(loc='best', shadow=True)
pyplot.xlabel(r"Some X")
pyplot.ylabel(r"Some Y")
pyplot.legend(loc='best', shadow=True)
pyplot.savefig("matplotlib1.pdf", bbox_inches='tight')
pyplot.show()
```

Now let's plot the dataframe directly from Pandas. It's easier to do and puts gridlines on the plot.

```python
df.plot(x='x', y='Run 1', label='Run 1')
df.plot(x='x', y='Run 2', label='Run 2')
pyplot.legend(loc='best', shadow=True)
pyplot.xlabel(r"Some X")
pyplot.ylabel(r"Some Y")
pyplot.legend(loc='best', shadow=True)
pyplot.savefig("matplotlib2.pdf", bbox_inches='tight')
pyplot.show()
```

Pandas has a default rcparams that we can use. My system shows a monospace font, but the example on the Pandas site (http://pandas-docs.github.io/pandas-docs-travis/visualization.html) shows a nicer looking font.

```python
# Pandas style
pd.options.display.mpl_style = 'default'

df.plot(x='x', y='Run 1', label='Run 1')
df.plot(x='x', y='Run 2', label='Run 2')
pyplot.legend(loc='best', shadow=True)
pyplot.xlabel(r"Some X")
pyplot.ylabel(r"Some Y")
pyplot.legend(loc='best', shadow=True)
pyplot.savefig("matplotlib3.pdf", bbox_inches='tight')

# Reset back to default matplotlib
plt.rcParams.update(rcdef)
```

### Plot in Gnuplot.py

gnuplot is a nice standalone plotting package, and Gnuplot.py provides a Python wrapper around it. The plots do not display inline but are rather output to a PostScript file. Their website is http://gnuplot-py.sourceforge.net/.

```python
import Gnuplot, Gnuplot.funcutils
```

```python
g = Gnuplot.Gnuplot(debug=1)
#g('set term postscript enhanced')
g('set key box')
g('set key bottom left')
g('set format y "%2.1f"')
g1 = Gnuplot.Data(xaxis, run1, title='Run 1', with_='lines linestyle 1 lw 2 lc rgb "blue"')
g2 = Gnuplot.Data(xaxis, run2, title='Run 2', with_='lines linestyle 1 lw 2 lc rgb "red"')
g.xlabel('Some X')
g.ylabel('Some Y')
g.plot(g1, g2)
g.hardcopy('gnuplot.eps', terminal='postscript', enhanced=1, color=1)
g.reset()
```

### Style from http://messymind.net/making-matplotlib-look-like-ggplot/

There are multiple attempts out there to try to get matplotlib to look like R's ggplot2. This code does a pretty good job. Getting matplotlib to look like R's output is a popular Stack Overflow question: http://stackoverflow.com/questions/14349055/making-matplotlib-graphs-look-like-r-by-default

```python
def rstyle(ax): 
    """Styles an axes to appear like ggplot2
    Must be called after all plot and axis manipulation operations have been carried out (needs to know final tick spacing)
    """
    #set the style of the major and minor grid lines, filled blocks
    ax.grid(True, 'major', color='w', linestyle='-', linewidth=1.4)
    ax.grid(True, 'minor', color='0.92', linestyle='-', linewidth=0.7)
    ax.patch.set_facecolor('0.85')
    ax.set_axisbelow(True)
    
    #set minor tick spacing to 1/2 of the major ticks
    ax.xaxis.set_minor_locator(MultipleLocator( (plt.xticks()[0][1]-plt.xticks()[0][0]) / 2.0 ))
    ax.yaxis.set_minor_locator(MultipleLocator( (plt.yticks()[0][1]-plt.yticks()[0][0]) / 2.0 ))
    
    #remove axis border
    for child in ax.get_children():
        if isinstance(child, matplotlib.spines.Spine):
            child.set_alpha(0)
       
    #restyle the tick lines
    for line in ax.get_xticklines() + ax.get_yticklines():
        line.set_markersize(5)
        line.set_color("gray")
        line.set_markeredgewidth(1.4)
    
    #remove the minor tick lines    
    for line in ax.xaxis.get_ticklines(minor=True) + ax.yaxis.get_ticklines(minor=True):
        line.set_markersize(0)
    
    #only show bottom left ticks, pointing out of axis
    rcParams['xtick.direction'] = 'out'
    rcParams['ytick.direction'] = 'out'
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    
    
    if ax.legend_ <> None:
        lg = ax.legend_
        lg.get_frame().set_linewidth(0)
        lg.get_frame().set_alpha(0.5)
        
        
def rhist(ax, data, **keywords):
    """Creates a histogram with default style parameters to look like ggplot2
    Is equivalent to calling ax.hist and accepts the same keyword parameters.
    If style parameters are explicitly defined, they will not be overwritten
    """
    
    defaults = {
                'facecolor' : '0.3',
                'edgecolor' : '0.28',
                'linewidth' : '1',
                'bins' : 100
                }
    
    for k, v in defaults.items():
        if k not in keywords: keywords[k] = v
    
    return ax.hist(data, **keywords)


def rbox(ax, data, **keywords):
    """Creates a ggplot2 style boxplot, is eqivalent to calling ax.boxplot with the following additions:
    
    Keyword arguments:
    colors -- array-like collection of colours for box fills
    names -- array-like collection of box names which are passed on as tick labels

    """

    hasColors = 'colors' in keywords
    if hasColors:
        colors = keywords['colors']
        keywords.pop('colors')
        
    if 'names' in keywords:
        ax.tickNames = plt.setp(ax, xticklabels=keywords['names'] )
        keywords.pop('names')
    
    bp = ax.boxplot(data, **keywords)
    pylab.setp(bp['boxes'], color='black')
    pylab.setp(bp['whiskers'], color='black', linestyle = 'solid')
    pylab.setp(bp['fliers'], color='black', alpha = 0.9, marker= 'o', markersize = 3)
    pylab.setp(bp['medians'], color='black')
    
    numBoxes = len(data)
    for i in range(numBoxes):
        box = bp['boxes'][i]
        boxX = []
        boxY = []
        for j in range(5):
          boxX.append(box.get_xdata()[j])
          boxY.append(box.get_ydata()[j])
        boxCoords = zip(boxX,boxY)
        
        if hasColors:
            boxPolygon = Polygon(boxCoords, facecolor = colors[i % len(colors)])
        else:
            boxPolygon = Polygon(boxCoords, facecolor = '0.95')
            
        ax.add_patch(boxPolygon)
    return bp
```

```python
# The plotting is a little different here, because you have to specify the axis to use the above code.
fig = pyplot.figure()
ax = fig.add_subplot(111)
pyplot.plot(xaxis, run1, label='Run 1')
pyplot.plot(xaxis, run2, label='Run 2')
ax.legend(loc='best', shadow=True)
pyplot.xlabel(r"Some X")
pyplot.ylabel(r"Some Y")
ax.legend(loc='best')
rstyle(ax)
plt.show()
fig.savefig("style1.pdf", bbox_inches='tight')
```
